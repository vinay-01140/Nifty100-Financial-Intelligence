```python
import re
import pandas as pd


def normalize_ticker(value):
    if value is None:
        return None

    value = str(value).strip().upper()
    value = value.replace(".NS", "")
    value = value.replace(".BO", "")
    value = value.replace("\n", "")
    value = value.replace("\t", "")
    value = re.sub(r"\s+", "", value)

    return value


companies = pd.read_excel("../data/raw/companies.xlsx", header=1)
companies["id"] = companies["id"].apply(normalize_ticker)
valid_ids = set(companies["id"].dropna())

tables = {
    "profitandloss": pd.read_excel("../data/raw/profitandloss.xlsx", header=1),
    "balancesheet": pd.read_excel("../data/raw/balancesheet.xlsx", header=1),
    "cashflow": pd.read_excel("../data/raw/cashflow.xlsx", header=1),
    "analysis": pd.read_excel("../data/raw/analysis.xlsx", header=1),
    "documents": pd.read_excel("../data/raw/documents.xlsx", header=1),
    "prosandcons": pd.read_excel("../data/raw/prosandcons.xlsx", header=1),
    "financial_ratios": pd.read_excel("../data/supporting/financial_ratios.xlsx"),
}

for name, df in tables.items():
    df["company_id_norm"] = df["company_id"].apply(normalize_ticker)
    invalid = sorted(set(df["company_id_norm"].dropna()) - valid_ids)

    print(f"\n{name}")
    print("invalid company ids count:", len(invalid))
    print(invalid[:50])
```

```python
import sqlite3
conn = sqlite3.connect("../nifty100.db")
```

```python
review_companies = pd.read_sql("""
SELECT id, company_name
FROM companies
ORDER BY id
LIMIT 5;
""", conn)

review_companies
```

```python
query = """
SELECT 
    c.id,
    c.company_name,

    (SELECT COUNT(*) FROM profitandloss p WHERE p.company_id = c.id) AS pnl_years,
    (SELECT COUNT(*) FROM balancesheet b WHERE b.company_id = c.id) AS bs_years,
    (SELECT COUNT(*) FROM cashflow cf WHERE cf.company_id = c.id) AS cf_years,
    (SELECT COUNT(*) FROM documents d WHERE d.company_id = c.id) AS document_count

FROM companies c
WHERE c.id IN ({})
ORDER BY c.id;
""".format(",".join([f"'{cid}'" for cid in review_companies["id"].tolist()]))

coverage_df = pd.read_sql(query, conn)
coverage_df
```

```python
less_than_5 = pd.read_sql("""
SELECT 
    c.id,
    c.company_name,
    COALESCE(p.pnl_years, 0) AS pnl_years,
    COALESCE(b.bs_years, 0) AS bs_years,
    COALESCE(cf.cf_years, 0) AS cf_years
FROM companies c
LEFT JOIN (
    SELECT company_id, COUNT(*) AS pnl_years
    FROM profitandloss
    GROUP BY company_id
) p ON c.id = p.company_id
LEFT JOIN (
    SELECT company_id, COUNT(*) AS bs_years
    FROM balancesheet
    GROUP BY company_id
) b ON c.id = b.company_id
LEFT JOIN (
    SELECT company_id, COUNT(*) AS cf_years
    FROM cashflow
    GROUP BY company_id
) cf ON c.id = cf.company_id
WHERE COALESCE(p.pnl_years, 0) < 5
   OR COALESCE(b.bs_years, 0) < 5
   OR COALESCE(cf.cf_years, 0) < 5
ORDER BY c.id;
""", conn)

less_than_5
```

```python
pnl_raw = pd.read_excel("../data/raw/profitandloss.xlsx", header=1)
bs_raw = pd.read_excel("../data/raw/balancesheet.xlsx", header=1)
cf_raw = pd.read_excel("../data/raw/cashflow.xlsx", header=1)

flagged_ids = ["ATGL", "AGTL", "JIOFIN", "SBIN"]

for cid in flagged_ids:
    print(f"\n================ {cid} ================")

    pnl_rows = pnl_raw[pnl_raw["company_id"].astype(str).str.upper() == cid]
    bs_rows = bs_raw[bs_raw["company_id"].astype(str).str.upper() == cid]
    cf_rows = cf_raw[cf_raw["company_id"].astype(str).str.upper() == cid]

    print("P&L rows:", len(pnl_rows))
    if len(pnl_rows) > 0:
        print(pnl_rows[["company_id", "year"]].head(10))

    print("BS rows:", len(bs_rows))
    if len(bs_rows) > 0:
        print(bs_rows[["company_id", "year"]].head(10))

    print("CF rows:", len(cf_rows))
    if len(cf_rows) > 0:
        print(cf_rows[["company_id", "year"]].head(10))
```
